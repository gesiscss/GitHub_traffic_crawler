# # This file is for presenting the data in tabular form

```python
from DataRetrieval import csvManipulation as csv
from IPython.core.display import display, HTML
```

```python
pandaFilesViews = csv.giveFullPandaFiles(type="views")
pandaFilesReferrers = csv.giveFullPandaFiles(type="referrers")
pandaFilesClones = csv.giveFullPandaFiles(type="referrers")
```

After the info is being crawled, we can present it in tabular fashion:

```python
for name, value in pandaFilesViews:
    display(HTML('<h1>'+name+'</h1>'))
    display(value)
```

```python
for name, value in pandaFilesReferrers:
    display(HTML('<h1>'+name+'</h1>'))
    display(value)
```

```python
for name, value in pandaFilesClones:
    display(HTML('<h1>'+name+'</h1>'))
    display(value)
```
